```python
# import pandas and matplotlib (pyplot for data visualization)
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# extract the data from the countries.csv file 
data = pd.read_csv('countries.csv')
data
```

```python
# extract the India data 
india_data = data[data.country == 'India']
india_data
```

```python
# extract the China data 
china_data = data[data.country == 'China']
china_data
```

```python
# extract the Japan data 
japan_data = data[data.country == 'Japan']
japan_data
```

```python
# extract the Indonesia data 
indonesia_data = data[data.country == 'Indonesia']
indonesia_data
```

```python
# extract the North Korea data
north_korea_data = data[data.country == 'Korea, Dem. Rep.']
north_korea_data
```

```python
# extract the South Korea data 
south_korea_data = data[data.country == 'Korea, Rep.']
south_korea_data
```

```python
# extract the Vietnam data 
vietnam_data = data[data.country == 'Vietnam']
vietnam_data
```

```python
# extract the Thailand data 
thailand_data = data[data.country == 'Thailand']
thailand_data
```

```python
# plot the absolute population growth of the Asian countries 
base = 10 ** 6
plt.plot(india_data.year, india_data.population / base)
plt.plot(china_data.year, china_data.population / base)
plt.plot(japan_data.year, japan_data.population / base)
plt.plot(north_korea_data.year, north_korea_data.population / base)
plt.plot(south_korea_data.year, south_korea_data.population / base)
plt.plot(vietnam_data.year, vietnam_data.population / base)
plt.plot(thailand_data.year, thailand_data.population / base)
plt.title('Absolute Population Growth')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(['India', 'China', 'Japan', 'North Korea', 'South Korea', 'Vietnam', 'Thailand'])
plt.show()
```

```python
# plot the percentage (relative to 1952) change of the population of the Asian countries
plt.plot(india_data.year, india_data.population / india_data.population.iloc[0] * 100)
plt.plot(china_data.year, china_data.population / china_data.population.iloc[0] * 100)
plt.plot(japan_data.year, japan_data.population / japan_data.population.iloc[0] * 100)
plt.plot(north_korea_data.year, north_korea_data.population / north_korea_data.population.iloc[0] * 100)
plt.plot(south_korea_data.year, south_korea_data.population / south_korea_data.population.iloc[0] * 100)
plt.plot(vietnam_data.year, vietnam_data.population / vietnam_data.population.iloc[0] * 100)
plt.plot(thailand_data.year, thailand_data.population / thailand_data.population.iloc[0] * 100)
plt.title('Percentage Change of Population (Rel. 1952)')
plt.xlabel('Year')
plt.ylabel('Population (% Rel. 1952)')
plt.legend(['India', 'China', 'Japan', 'North Korea', 'South Korea', 'Vietnam', 'Thailand'])
plt.show()
```

```python
# conclusions 
print('From the graphs above, we can conclude the following about the Asian countries')
print('1. Vietnam and Thailand have sported the highest population growth rates, but '
     + 'some of the lowest populations')
print('2. India and China sport the highest populations')
print('3. Japan, North Korea, South Korea, Vietnam, and Thailand sport '
     + 'the lowest populations')
```
