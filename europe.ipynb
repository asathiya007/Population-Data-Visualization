```python
# import pandas and matplotlib (pyplot) for data visualization 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# extract the data from the countries.csv file
data = pd.read_csv('countries.csv')
data
```

```python
# extract the UK data 
uk = data[data.country == 'United Kingdom']
uk
```

```python
# extract the France data 
france = data[data.country == 'France']
france
```

```python
# extract the Germany data
germany = data[data.country == 'Germany']
germany
```

```python
# extract the Spain data 
spain = data[data.country == 'Spain']
spain
```

```python
# extract the Portugal data 
portugal = data[data.country == 'Portugal']
portugal
```

```python
# plot the absolute population growth of the European countries 
unit = 10 ** 6
plt.plot(uk.year, uk.population / unit)
plt.plot(france.year, france.population / unit)
plt.plot(germany.year, germany.population / unit)
plt.plot(spain.year, spain.population / unit)
plt.plot(portugal.year, portugal.population / unit)
plt.legend(['United Kingdom', 'France', 'Germany', 'Spain', 'Portugal'])
plt.xlabel('Year')
plt.ylabel('Population (In Millions)')
plt.title('Absolute Population Growth')
plt.show()
```

```python
# plot the percentage population growth of the European countries 
plt.plot(uk.year, uk.population / uk.population.iloc[0] * 100)
plt.plot(france.year, france.population / france.population.iloc[0] * 100)
plt.plot(germany.year, germany.population / germany.population.iloc[0] * 100)
plt.plot(spain.year, spain.population / spain.population.iloc[0] * 100)
plt.plot(portugal.year, portugal.population / portugal.population.iloc[0] * 100)
plt.legend(['United Kingdom', 'France', 'Germany', 'Spain', 'Portugal'])
plt.xlabel('Year')
plt.ylabel('Population (% Rel. 1952)')
plt.title('Percentage Change of Population (Rel. 1952)')
plt.show()
```

```python
# conclusions 
print('From the graphs above, we can conclude the following about the European countries:')
print('1. Germany sports the highest population and Portugal sports the lowest population')
print('2. Spain and France sport similar population growth rates and the UK, Portugal, and Germany sport similar population growth rates')
print('3. Spain and France sport a higher population growth rate than Germany, Spain, and Portugal')
```
