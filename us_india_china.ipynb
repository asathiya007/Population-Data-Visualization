```python
# import pandas and matplotlib (pyplot) for data visualization 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# extract the data from the countries.csv file 
data = pd.read_csv('countries.csv')
data
```

```python
# extract the US data 
us_data = data[data.country == 'United States']
us_data
```

```python
# extract the India data 
india_data = data[data.country == 'India']
india_data
```

```python
# extract the China data 
china_data = data[data.country == 'China']
china_data
```

```python
# plot the absolute population growth of China and the US 
plt.plot(us_data.year, us_data.population / 10**6)
plt.plot(india_data.year, india_data.population / 10 ** 6)
plt.plot(china_data.year, china_data.population / 10**6)
plt.title('Absolute Population Growth')
plt.xlabel('Year')
plt.ylabel('Population (In Millions)')
plt.legend(['United States', 'India', 'China'])
plt.show()
```

```python
# plot the percentage (relative to 1952) change of the population of the US, India, and China
plt.plot(us_data.year, us_data.population / us_data.population.iloc[0] * 100)
plt.plot(india_data.year, india_data.population / india_data.population.iloc[0] * 100)
plt.plot(china_data.year, china_data.population / china_data.population.iloc[0] * 100)
plt.title('Percentage Change of Population (Rel. 1952)')
plt.xlabel('Year')
plt.ylabel('Population (% Rel. 1952)')
plt.legend(['United States', 'India', 'China'])
plt.show()
```

```python
# conclusions 
print('From the graphs above, we can conclude the following about the US, India, and China: ')
print('1. China sports the highest population')
print('2. India sports the highest population growth rate')
print('3. The US sports the lowest population and the lowest population growth rate')
```
